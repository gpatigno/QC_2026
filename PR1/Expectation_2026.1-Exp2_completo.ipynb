```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import SparsePauliOp, Statevector
from qiskit.visualization import plot_bloch_multivector
from qiskit.visualization import array_to_latex

import numpy as np

from matplotlib import pyplot as plt

%matplotlib inline
```

```python
def pauli_z_on_qubit(i, n):
    """
    Construye el string de Pauli correspondiente al operador Z_i
    respetando el orden little-endian de Qiskit.
    """

    pauli = ["I"] * n
    pauli[n - i - 1] = "Z"   # conversión al orden little-endian

    return "".join(pauli)
```

```python
def expectation_per_qubit(circuito):
    # Obtener estado cuántico
    state = Statevector.from_instruction(circuito)
    expectations = []

    # Calcular <Z_i> para cada qubit
    for i in range(circuito.num_qubits):

        pauli_string = pauli_z_on_qubit(i, circuito.num_qubits)

        observable = SparsePauliOp(pauli_string)

        exp = float(np.real(state.expectation_value(observable)))
        expectations.append(exp)

    return expectations
```

```python
def circuitoQ(n):

    # Crear circuito
    qc = QuantumCircuit(n)

    # Preparar estado |+>^n
    for i in range(n):
        qc.ry(np.pi / 3, i)

    return qc
```

```python
circuito = circuitoQ(1)
```

```python
estado = Statevector(circuito)
display(array_to_latex(estado, prefix="\\ket{\\psi_{out}} = "))
```

```python
plot_bloch_multivector(estado)   
plt.show()
```

```python
fig = plot_bloch_multivector(estado)   
ax = fig.axes[0]
# change camera angle
ax.view_init(azim=-90, elev=0)

plt.show()
```

```python
n = 5
circuito = circuitoQ(n)
```

```python
ValorEsperado = expectation_per_qubit(circuito)
print(ValorEsperado)
```
