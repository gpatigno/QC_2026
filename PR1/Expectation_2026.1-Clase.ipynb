```python
def pauli_z_on_qubit(i, n):
    """
    Construye el string de Pauli correspondiente al operador Z_i
    respetando el orden little-endian de Qiskit.
    """

    pauli = ["I"] * n
    pauli[n - i - 1] = "Z"   # conversión al orden little-endian

    return "".join(pauli)
```
