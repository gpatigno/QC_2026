```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector, SparsePauliOp
from qiskit.visualization import array_to_latex
```

```python
# Crear circuito
qc = QuantumCircuit(1)
qc.h(0)   # Crear superposición
```

```python
# Obtener el estado final
state = Statevector.from_instruction(qc)
```

```python
print("Estado final:")
array_to_latex(state, prefix="\\ket{\\psi_0} = ")
```

```python
# Definir observable (Pauli Z)
observable = SparsePauliOp.from_list([("Z", 1)])
```

```python
# Calcular valor esperado
expectation_value = state.expectation_value(observable)
```

```python
print("\nValor esperado ⟨Z⟩:")
print(expectation_value.real)
```
